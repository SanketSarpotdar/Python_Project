```python
# 1] What is the distribution of customers by age group?
plt.figure(figsize=(10, 5))
sb.histplot(df['age'], bins=20, kde=True, color='blue')
plt.title("Distribution of Customer Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 5))
sb.boxplot(x=df['age'], color='blue')
plt.title("Box Plot of Customer Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
# 2] What is the average income of customers by gender?
avg_income_gender = df.groupby('gender')['income'].mean()
plt.figure(figsize=(8, 4))
sb.barplot(x=avg_income_gender.index, y=avg_income_gender.values, palette='coolwarm')
plt.title("Average Income by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Income")
plt.show()

plt.figure(figsize=(8, 5))
sb.boxplot(x=df['gender'], y=df['income'], palette='coolwarm')
plt.title("Income Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Income")
plt.show()
```

```python
# 3] What is the total revenue generated by each product category?
total_revenue_category = prd1.groupby('category')['total_amount'].sum()
plt.figure(figsize=(10, 5))
sb.barplot(x=total_revenue_category.index, y=total_revenue_category.values, palette='plasma')
plt.title("Total Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()

average_revenue_category = prd1.groupby('category')['total_amount'].mean()

# Plot the Average Revenue per Order by Category
plt.figure(figsize=(10, 5))
sb.barplot(x=average_revenue_category.index, y=average_revenue_category.values, palette='viridis')
plt.title("Average Revenue per Order by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Revenue per Order")
plt.xticks(rotation=45)
plt.show()
```

```python
# 4] What are the top 10 selling products by total revenue?
top_products = prd1.groupby('product_id')['total_amount'].sum().nlargest(10)
plt.figure(figsize=(10, 5))
sb.barplot(x=top_products.index, y=top_products.values, palette='magma')
plt.title("Top 10 Selling Products by Revenue")
plt.xlabel("Product ID")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()


top_products_qty = prd1.groupby('product_id')['quantity'].sum().nlargest(10)
plt.figure(figsize=(10, 5))
sb.barplot(x=top_products_qty.index, y=top_products_qty.values, palette='cividis')
plt.title("Top 10 Selling Products by Quantity Sold")
plt.xlabel("Product ID")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5] What is the average purchase value for each product category?
# Count the total number of unique product categories
total_categories = prd1['category'].nunique()
print("Total number of unique product categories:", total_categories)

category_avg_purchase = prd1.groupby('category')['total_amount'].mean()
plt.figure(figsize=(10, 5))
sb.barplot(x=category_avg_purchase.index, y=category_avg_purchase.values, palette='Blues')
plt.title("Average Purchase Value by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Purchase Value")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(9, 4))
sb.boxplot(x='category', y='total_amount', data=prd1, palette='coolwarm')
plt.title("Distribution of Purchase Value by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Purchase Value")
plt.xticks(rotation=45)
plt.show()
```

```python
# 6] What is the distribution of purchase amounts?
plt.figure(figsize=(10, 5))
sb.histplot(prd1['total_amount'], bins=30, kde=True, color='blue')
plt.title("Distribution of Purchase Amounts")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 4))
sb.boxplot(x=prd1['total_amount'], color='green')
plt.title("Box Plot of Purchase Amounts")
plt.xlabel("Purchase Amount")
plt.show()
```

```python
# 7] What is the correlation between income and total spent?
plt.figure(figsize=(8, 6))
sb.scatterplot(x=cd['income'], y=pfd['total_spent'], alpha=0.6)
plt.title("Correlation between Income and Total Spent")
plt.xlabel("Income")
plt.ylabel("Total Spent")
plt.show()

plt.figure(figsize=(8, 5))
plt.hexbin(cd['income'], pfd['total_spent'], gridsize=30, cmap='Blues', alpha=0.8)
plt.colorbar(label="Number of Customers")
plt.title("Income vs. Total Spent Density Heatmap")
plt.xlabel("Income")
plt.ylabel("Total Spent")
plt.show()
```

```python
# 8] What are the top brands by total revenue?
top_brands = prd1.merge(prd, on='product_id').groupby('brand')['total_amount'].sum().nlargest(10)
plt.figure(figsize=(10, 5))
sb.barplot(x=top_brands.index, y=top_brands.values, palette='coolwarm')
plt.title("Top Brands by Total Revenue")
plt.xlabel("Brand")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()

top_brands_qty = prd1.merge(prd, on='product_id').groupby('brand')['quantity'].sum().nlargest(10)
plt.figure(figsize=(6, 6))
plt.pie(top_brands_qty.values, labels=top_brands_qty.index, autopct='%1.1f%%', colors=sb.color_palette('pastel'))
plt.title("Top Brands by Quantity Sold")
plt.show()
```

```python
# 9] How does average purchase value vary by customer age group?
age_bins = [18, 25, 35, 45, 55, 65, 100]
cd['age_group'] = pd.cut(cd['age'], bins=age_bins, labels=["18-24", "25-34", "35-44", "45-54", "55-64", "65+"])
merged_data = cd.merge(pfd, on='customer_id')
plt.figure(figsize=(10, 5))
sb.boxplot(x='age_group', y='avg_purchase_value', data=merged_data, palette='coolwarm')
plt.title("Average Purchase Value by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Purchase Value")
plt.show()

# plt.figure(figsize=(8, 5))
# sb.violinplot(x='age_group', y='avg_purchase_value', data=merged_data, palette='viridis', inner='quartile')
# plt.title("Distribution of Average Purchase Value by Age Group")
# plt.xlabel("Age Group")
# plt.ylabel("Average Purchase Value")
# plt.show()
```
